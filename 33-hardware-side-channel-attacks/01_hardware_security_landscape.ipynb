# 01 — Hardware Security Landscape

## What This Is
Hardware security encompasses attacks and defences at the physical layer: side-channel attacks (power, timing, EM), fault injection (voltage glitching, laser), hardware trojans, and supply chain integrity. These attacks bypass software security entirely.

## Why It Matters
Software defences — encryption, authentication, sandboxing — all rely on correct hardware behaviour. Spectre/Meltdown showed that microarchitectural attacks can leak secrets across trust boundaries. Hardware roots of trust (HSM, TPM, Secure Enclave) are foundational to modern security.

## Where the Community Stands
ChipWhisperer is the standard open-source platform for power analysis and glitching. CHES (Conference on Cryptographic Hardware and Embedded Systems) is the premier academic venue. Industry uses Common Criteria and FIPS 140-3 certifications to evaluate hardware security.

## Attack Taxonomy
- **Passive attacks**: observe side channels without disturbing the device
- **Active attacks**: inject faults to cause incorrect computation
- **Invasive attacks**: physically alter the chip (decapping, probing)
- **Semi-invasive**: interact without full decap (laser fault injection)

## Hardware Security Evaluation Frameworks
Common Criteria (ISO 15408) defines Evaluation Assurance Levels (EAL1-7). FIPS 140-3 defines four security levels for cryptographic modules. These certifications are required for government procurement and payment systems.

```python
EAL_LEVELS = {
    1: ('Functionally Tested',        'Basic — documentation reviewed'),
    2: ('Structurally Tested',        'Vulnerability analysis + basic tests'),
    3: ('Methodically Tested',        'Development controls + automated testing'),
    4: ('Methodically Designed',      'Full design + independent testing — most common commercial'),
    5: ('Semiformally Designed',      'Formal design methods + covert channel analysis'),
    6: ('Semiformally Verified',      'Near-formal verification — smartcards, HSMs'),
    7: ('Formally Verified',          'Full mathematical proof — rare, ultra-high security'),
}

FIPS_140_LEVELS = {
    1: 'Basic — software only, no physical protection',
    2: 'Tamper-evident coating or seals required',
    3: 'Tamper-responsive (zeroises keys on intrusion)',
    4: 'Complete physical envelope protection (HSM)',
}

print('Common Criteria EAL Levels:')
for level, (name, desc) in EAL_LEVELS.items():
    print(f'  EAL{level}: {name} — {desc}')

print('\nFIPS 140-3 Security Levels:')
for level, desc in FIPS_140_LEVELS.items():
    print(f'  Level {level}: {desc}')
```
