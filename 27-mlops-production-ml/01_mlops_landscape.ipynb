# 01 — MLOps Landscape

## What
MLOps (Machine Learning Operations) is the set of practices that unify ML system development (Dev) and ML system operation (Ops). Unlike traditional software, ML systems have an additional axis of concern: *data* and *model* quality decay over time even when code stays the same.

## Why
Research shows that only a small fraction of ML systems in production are ML code — the rest is configuration, data collection, feature engineering, monitoring, and serving infrastructure. Without MLOps discipline, models silently degrade, experiments are unreproducible, and teams waste months re-doing work.

## Community context
The MLOps maturity model (Google, Microsoft, AWS) defines three levels: (0) manual, script-based pipelines; (1) ML pipeline automation; (2) CI/CD pipeline automation. Most organisations sit at level 0. Tools like MLflow, Weights & Biases, DVC, Kubeflow, and Seldon have emerged to bridge the gap.

```python
# ML System Audit Checklist
# A practical tool to assess where your team sits on the MLOps maturity ladder

audit_questions = [
    ('Data', 'Is training data versioned and reproducible?'),
    ('Data', 'Is there automated validation for data schema and distributions?'),
    ('Data', 'Are training/serving feature pipelines identical?'),
    ('Experiments', 'Are all experiments tracked with hyperparameters and metrics?'),
    ('Experiments', 'Can any past experiment be exactly reproduced?'),
    ('Models', 'Are model artifacts versioned in a registry?'),
    ('Models', 'Is there a formal model promotion workflow (staging -> prod)?'),
    ('Pipeline', 'Are training pipelines automated and triggered on data changes?'),
    ('Pipeline', 'Is there a CI pipeline that trains, evaluates, and tests models?'),
    ('Serving', 'Are model predictions logged for offline analysis?'),
    ('Serving', 'Is latency monitored per endpoint?'),
    ('Monitoring', 'Is input distribution drift detected automatically?'),
    ('Monitoring', 'Are prediction distribution shifts alerted on?'),
    ('Retraining', 'Is retraining triggered automatically on drift or performance drops?'),
    ('Governance', 'Are model cards and datasheets maintained for all prod models?'),
]

def run_audit(answers: dict):
    scores = {}
    for domain, q in audit_questions:
        scores.setdefault(domain, [])
        scores[domain].append(1 if answers.get(q, False) else 0)
    print('=== MLOps Maturity Audit ===')
    total = 0
    for domain, vals in scores.items():
        pct = 100*sum(vals)//len(vals)
        bar = '#'*(pct//10) + '-'*(10-pct//10)
        print(f'  {domain:<14} [{bar}] {pct}%')
        total += sum(vals)
    overall = 100*total//len(audit_questions)
    level = 0 if overall < 35 else (1 if overall < 70 else 2)
    print(f'\nOverall: {overall}% -> MLOps Level {level}')
    return overall

# Example: a team with basic experiment tracking but no automation
sample_answers = {q: False for _, q in audit_questions}
sample_answers['Are all experiments tracked with hyperparameters and metrics?'] = True
sample_answers['Can any past experiment be exactly reproduced?'] = True
sample_answers['Are model artifacts versioned in a registry?'] = True
run_audit(sample_answers)
```
