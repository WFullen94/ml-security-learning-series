# 01 — Multimodal AI Landscape

## What
Multimodal AI models process and reason across multiple modalities — text, images, audio, video, and structured data — either jointly or through cross-modal alignment. The field has exploded since CLIP (2021), driven by the observation that rich representations emerge when models are trained on aligned multi-modal data at scale.

## Why
Unimodal models are bottlenecked by the modality of their input. Vision-language models unlock tasks like visual question answering, image captioning, chart understanding, and document intelligence that cannot be solved by text-only or vision-only systems. The most capable frontier models (GPT-4V, Gemini, Claude 3+) are all multimodal.

## Community context
Two architectural paradigms dominate: (1) *contrastive alignment* — CLIP-style models that learn a shared embedding space via InfoNCE loss; (2) *generative fusion* — models like Flamingo, LLaVA, and GPT-4V that inject visual tokens into a language model's context. BLIP-2 and InstructBLIP add a lightweight Q-Former bridge between vision encoder and LLM. Retrieval-augmented multimodal generation (RAMM) extends this with external knowledge.

## Key Architectural Patterns

1. **Dual encoder (CLIP-style)**: Encode each modality separately, align in shared embedding space via contrastive loss.
2. **Encoder-decoder fusion (Flamingo-style)**: Freeze a vision encoder, inject visual tokens into language model via cross-attention gates.
3. **Q-Former bridge (BLIP-2-style)**: Learnable query tokens attend to image features, compressing them to a fixed-length representation for the LLM.
4. **Unified tokenisation (Gemini/GPT-4o-style)**: All modalities are tokenised into a single discrete vocabulary that the transformer processes natively.
