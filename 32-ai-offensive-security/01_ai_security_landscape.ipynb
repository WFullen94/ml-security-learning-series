# 01 — AI-Assisted Offensive Security Landscape

## What
This series covers AI-assisted security techniques in the context of **authorized penetration testing, CTF competitions, and defensive research**. AI tools are being used both by defenders (threat detection, anomaly detection, vulnerability scanning) and by red teams (accelerating reconnaissance, generating payloads, automating vulnerability discovery).

**All content in this series applies exclusively to authorized engagements.** Unauthorized access to systems is illegal under the Computer Fraud and Abuse Act (CFAA), UK Computer Misuse Act, and equivalents worldwide.

## Why
Understanding offensive techniques is prerequisite to building effective defenses. Blue teams that understand how attackers think — and what AI accelerates — can prioritize detection capabilities and harden systems more effectively. The 'know your adversary' principle is foundational to NIST CSF and MITRE ATT&CK.

## Community context
MITRE ATT&CK is the standard taxonomy of adversary tactics, techniques, and procedures (TTPs). AI is being integrated into: (1) automated vulnerability scanning (GPT-powered Nuclei templates); (2) phishing generation at scale; (3) code analysis for vulnerability discovery; (4) C2 framework evasion via ML-generated network patterns. Defenders are countering with AI-powered SIEM rules, NLP for log analysis, and ML-based anomaly detection.

```python
# MITRE ATT&CK framework: AI-relevant technique mapping
import json

# Subset of ATT&CK techniques where AI is having the most impact
ai_impact_techniques = {
    'Reconnaissance': {
        'T1592 - Gather Victim Host Info': {
            'ai_impact': 'HIGH',
            'ai_application': 'LLMs parse OSINT faster, correlate across sources',
            'defensive_detection': 'Monitor for scraping patterns, rate-limit APIs'
        },
        'T1589 - Gather Victim Identity Info': {
            'ai_impact': 'HIGH',
            'ai_application': 'AI-generated spear-phishing based on LinkedIn/social data',
            'defensive_detection': 'User behaviour analytics, email authentication (DMARC)'
        },
    },
    'Initial Access': {
        'T1566 - Phishing': {
            'ai_impact': 'CRITICAL',
            'ai_application': 'LLM-generated personalised phishing at scale, no grammar errors',
            'defensive_detection': 'URL analysis, sandbox, security awareness training'
        },
        'T1190 - Exploit Public-Facing Application': {
            'ai_impact': 'MEDIUM',
            'ai_application': 'AI-assisted fuzzing, CVE-to-exploit mapping',
            'defensive_detection': 'WAF, patch management, vulnerability scanning'
        },
    },
    'Defense Evasion': {
        'T1027 - Obfuscated Files or Information': {
            'ai_impact': 'MEDIUM',
            'ai_application': 'ML-generated code variants that evade signature detection',
            'defensive_detection': 'Behavioural detection, sandboxing, EDR'
        },
    },
    'Exfiltration': {
        'T1048 - Exfiltration Over Alternative Protocol': {
            'ai_impact': 'MEDIUM',
            'ai_application': 'ML-shaped DNS/HTTPS tunnelling traffic to mimic normal',
            'defensive_detection': 'DNS anomaly detection, ML-based DLP'
        },
    },
}

print('AI Impact on MITRE ATT&CK Techniques (authorized research context):')
for tactic, techniques in ai_impact_techniques.items():
    print(f'\n[{tactic}]')
    for technique, info in techniques.items():
        print(f'  {technique}')
        print(f'    AI Impact:  {info["ai_impact"]}')
        print(f'    Offense:    {info["ai_application"]}')
        print(f'    Defense:    {info["defensive_detection"]}')
```
