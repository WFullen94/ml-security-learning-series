# 01 — AI Safety Landscape

A survey of technical AI safety: what we're trying to prevent, and how.

## Alignment Tax, Inner vs Outer Alignment

**Outer alignment**: does the specified objective capture what we actually want? Goodhart's Law applies: 'When a measure becomes a target, it ceases to be a good measure.'
- Example: train a game agent to maximise score → it finds unintended exploits
- Example: optimise for human approval → model learns to be sycophantic

**Inner alignment** (Hubinger et al., 2019): even if the objective is correct, does the trained model actually pursue it?
- Mesa-optimisation: the trained model may itself be an optimiser with different goals than the training objective
- A model can 'appear' to pursue the objective during training but have a different internal goal that manifests only in deployment

**The alignment tax**: safety-focused training typically reduces raw capability metrics. This creates pressure to skip safety work, especially in competitive environments.

**Current technical approaches**:
- *RLHF/Constitutional AI*: align to human preferences via reinforcement learning from human feedback
- *Scalable oversight*: maintain human understanding of model behaviour as models scale
- *Interpretability*: understand internal representations to verify alignment
- *Formal verification*: prove properties about model behaviour

```python
# Safety benchmark evaluation suite
import numpy as np
from collections import defaultdict

# Simulated safety evaluation framework
class SafetyEvaluator:
    def __init__(self):
        self.benchmarks = [
            'harmlessness',
            'helpfulness',
            'honesty',
            'sycophancy_resistance',
            'instruction_following',
            'robustness_to_jailbreaks',
        ]

    def evaluate_model(self, model_name, model_type='base'):
        """Simulate safety evaluation scores."""
        np.random.seed(hash(model_name) % 1000)

        if model_type == 'base':
            # Base models: capable but less safe
            scores = {
                'harmlessness': 0.55 + np.random.uniform(-0.1, 0.1),
                'helpfulness': 0.75 + np.random.uniform(-0.1, 0.1),
                'honesty': 0.70 + np.random.uniform(-0.1, 0.1),
                'sycophancy_resistance': 0.45 + np.random.uniform(-0.1, 0.1),
                'instruction_following': 0.80 + np.random.uniform(-0.05, 0.05),
                'robustness_to_jailbreaks': 0.40 + np.random.uniform(-0.1, 0.1),
            }
        elif model_type == 'rlhf':
            # RLHF models: safer but may be sycophantic
            scores = {
                'harmlessness': 0.85 + np.random.uniform(-0.05, 0.05),
                'helpfulness': 0.80 + np.random.uniform(-0.05, 0.05),
                'honesty': 0.75 + np.random.uniform(-0.05, 0.05),
                'sycophancy_resistance': 0.55 + np.random.uniform(-0.1, 0.1),
                'instruction_following': 0.90 + np.random.uniform(-0.05, 0.05),
                'robustness_to_jailbreaks': 0.70 + np.random.uniform(-0.05, 0.05),
            }
        else:  # constitutional_ai
            scores = {
                'harmlessness': 0.90 + np.random.uniform(-0.03, 0.03),
                'helpfulness': 0.82 + np.random.uniform(-0.05, 0.05),
                'honesty': 0.85 + np.random.uniform(-0.03, 0.03),
                'sycophancy_resistance': 0.70 + np.random.uniform(-0.05, 0.05),
                'instruction_following': 0.88 + np.random.uniform(-0.03, 0.03),
                'robustness_to_jailbreaks': 0.80 + np.random.uniform(-0.05, 0.05),
            }
        return scores

evaluator = SafetyEvaluator()

models = [
    ('Model-A-Base', 'base'),
    ('Model-A-RLHF', 'rlhf'),
    ('Model-A-CAI', 'constitutional_ai'),
]

print(f'{"Benchmark":<28}', end='')
for name, _ in models:
    print(f'{name:>15}', end='')
print()
print('-' * (28 + 15 * len(models)))

all_scores = {}
for name, mtype in models:
    all_scores[name] = evaluator.evaluate_model(name, mtype)

for benchmark in evaluator.benchmarks:
    print(f'{benchmark:<28}', end='')
    for name, _ in models:
        print(f'{all_scores[name][benchmark]:>15.3f}', end='')
    print()

print('\nOverall safety score (mean):')
for name, _ in models:
    mean = np.mean(list(all_scores[name].values()))
    print(f'  {name}: {mean:.3f}')
```
