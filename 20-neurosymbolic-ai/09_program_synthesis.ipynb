# Program Synthesis

## Program Synthesis

Program synthesis automatically generates programs that satisfy a specification. LLM-based synthesis generates code from natural language, which can then be formally verified.

**Sketch-based synthesis**: Provide a partial program (sketch) with holes; the synthesizer fills in the holes.
**Verified synthesis**: Generate code → run tests → if tests fail, use feedback to regenerate. Loop until all tests pass or budget exhausted.
**RLHF for code**: Fine-tune LLMs with human feedback on code quality — correctness, efficiency, readability.

```python
from typing import List, Dict, Optional, Callable
import ast

# Simplified program synthesis with verification

class ProgramSynthesizer:
    def __init__(self):
        # Template programs (simulating LLM generation)
        self.templates = [
            'def solve(nums): return sorted(nums)',
            'def solve(nums): return sorted(nums, reverse=True)',
            'def solve(nums): return [x for x in nums if x > 0]',
            'def solve(nums): return sum(nums)',
            'def solve(nums): return max(nums)',
        ]
    
    def generate_candidates(self, spec: str, n: int = 3) -> List[str]:
        import random
        return random.sample(self.templates, min(n, len(self.templates)))
    
    def verify_against_tests(self, code: str, tests: List[Dict]) -> bool:
        try:
            namespace = {}
            exec(compile(ast.parse(code), '<string>', 'exec'), namespace)
            fn = namespace.get('solve')
            if fn is None: return False
            return all(fn(t['input']) == t['expected'] for t in tests)
        except:
            return False

synthesizer = ProgramSynthesizer()

# Test cases
tasks = [
    {
        'spec': 'Sort a list of numbers in ascending order',
        'tests': [
            {'input': [3, 1, 4, 1, 5], 'expected': [1, 1, 3, 4, 5]},
            {'input': [9, 2, 7], 'expected': [2, 7, 9]},
        ]
    },
    {
        'spec': 'Return the sum of a list',
        'tests': [
            {'input': [1, 2, 3], 'expected': 6},
            {'input': [10, 20], 'expected': 30},
        ]
    },
]

print('Verified Program Synthesis:')
for task in tasks:
    candidates = synthesizer.generate_candidates(task['spec'])
    found = None
    for code in candidates:
        if synthesizer.verify_against_tests(code, task['tests']):
            found = code
            break
    print(f'  Spec: {task["spec"]}')
    if found:
        print(f'  Synthesized: {found}')
    else:
        print(f'  No verified program found in {len(candidates)} candidates')
    print()
```
